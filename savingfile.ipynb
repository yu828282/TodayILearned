```python
from PIL import ImageGrab
import keyboard

try : 
    while True :
        x, y = pyautogui.position()
        color = ImageGrab.grab().getpixel((x, y))
        print(f'x, y : ({x}, {y}, color : {color})')
        time.sleep(3)

        if keyboard.is_pressed('q'):
            print("'q' pressed. Exiting.....")
            break
            
except KeyboardInterrupt:
    print('end')
```

```python
from bs4 import BeautifulSoup

url = 'https://www.foxnews.com/' # request a fox news website
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'lxml') # pasing html with BeautifulSoup

headlines = soup.find_all('h3', class_= 'title') # finding html tag with h3, class title

#for i, headline in enumerate(headlines, start = 1):
#    print(f'{i}. {headline.text.strip()}')

with open('headlines.txt', 'w', encoding='utf-8') as file:
    for i, headline in enumerate(headlines, start = 1):
        file.write(f'{i}. {headline.text.strip()}\n')
        
print('finish')
```
